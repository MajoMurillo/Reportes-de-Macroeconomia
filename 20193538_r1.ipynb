# Pregunta 1
### Derive y explique la función de demanda de Consumo
Cuando hablamos de demanda de Consumo, nos referimos al gasto que hacen las familias y la distribucion del dinero en donde:

T = tributacion total, es decir, todo lo que el pais puede recaudar de nuestros impuestos
t = tasa impositiva, la cual varia de 0 (no paga impuestos) a 1 (paga mas impuestos de lo debido)

Gracias a estas dos variables, tendremos nuestra primera formula la cual sera

T = t * Y

Para hallar la funcion del consumo, tenemos que tomar en cuenta el ingreso disponible (Yd), ya que es el dinero que quedara a las familias sin los impuestos, cuya formula es:
Yd=Y−T

Entonces Yd = (1 − t) ∗ Y

Sabiendo el ingreso disponible, podemos hallar la funcion de consumo, la cual seria

C=Co⁡+b ∗ Yd

donde Co = consumo fijo y b = gastos variables

### Supuestos del modelo Ingreso-Gasto Keynesiano
Segun Felix Jimenez, en su libro "Apuntes sobre macroeconomia Intermedia" (pg.56) indica que el modelo de Ingreso-gasto keynesiano supone:

1- que el nivel de los precios es fijo en el corto plazo, es decir, si el contexto o las condiciones economicas cambian, el nivel de precio no se ajustara rapidamente a el.

2- que el mercado monetario no determina la tasa de interes, si no fuera de este.

3- que el nivel de producto se adapta a cualquier cambio en la demanda agregada

# Pregunta 4
### Encuentre el nivel de Ingreso de equilibrio (Ye) 
Tomemos en cuenta que el ingreso de equilibrio es exactamente igual a la demanda agregada de equilibrio ya que estan sobre un grafico de 45 grados, es decir DA = Ye

Donde:
Consumo = Co + bYd
Inversion= I = Io - hr
Gobierno= G = Go
Tributacion= T= tY
Exportaciones X=X0
Importaciones= M=mYd
Demanda agregada= DA=C+I+G+X-M

Entonces:
DA = C + I + G + X - M
DA = Co + b(1-t)Y + Io - hr + Go + Xo - m(1-t)Y
DA = Co + Io - hr + Go + Xo + b(1-t)Y - m(1-t)Y
DA= Co + Io - hr + Go + Xo + (b-m)(1-t)Y

Reducimos "Co + Io - hr + Go + Xo" a "α0" y "(b-m)(1-t)" a "α1", y nos quedaria de esta manera:

DA= α0 + α1 * Y

Al querer aplicarlo para el Ingreso de equilibrio (Ye)seria:
Ye = α0 + α1 * Y
Ye - (b-m)(1-t)Y = Co + Io - hr + Go + Xo
(1- (b-m)(1-t))Y = Co + Io - hr + Go + Xo
Factorizamos el Ye y finalmente nos quedaria

Ye= 1/1−(b−m)(1−t)∗{Co+Io+Go+Xo−hr}

en otras palabras: Ye = {1/1-α1}* {α0}

```python
Y_size = 100
Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9
Y = np.arange(Y_size)
```

```python
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
a = 2.5

def L_45(a, Y):
 L_45 = a*Y
 return L_45
L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5, ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=145, xmin= 0, xmax= 0.8, linestyle = ":", color = "grey")

# Texto agregado
plt.text(85, 192, '$DA=α_o + α_1Y$', fontsize = 11.5, color = 'black')
plt.text(82, 170, '$α_1=(b-m)(1-t) $', fontsize = 11.5, color = 'black')
plt.text(70, 110, '$α_o=(Co+Io+Go+Xo-hr) $', fontsize = 11.5, color = 'black')
plt.text(70, 130, '$↑$', fontsize = 18, color = 'black')
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

# línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda
plt.show()
```

# Pregunta 5
### Analice una Política Fiscal Expansiva con aumento del Gasto del Gobierno
Ocurre cuando el gobierno (Go) aumenta su gasto, destinandolo a impulsar el sector produccion y laboral en la economia.
=> Dado ese caso, las variables como Consumo (Co), Inversion (Io) y Exportaciones (Xo), no van a cambiar al ser exogenas

Al aumentar el gasto gubernamental (Go), aumenta tambien el intercepto de la funcion de Demanda Agregada (DA), la cual al ser igual al Interes de Equilibrio (Ye), hace que este tambien aumente

Segun Jimenez, F., matematicamente se puede expresar de la siguiente forma 
dY= {1/1−(b−m)(1−t)}∗{dCo+dIo+dGo+dXo−hrd}
Y como es mayor que 0, entonces se expresa: 
(dY/dG)=(1/1-(b-m)(1-t))>0

Graficamente, la DA se desplaza hacia arriba

```python
Y_size = 100
Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
Y = np.arange(Y_size)

 # Ecuación
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K
DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio
# Se define el parámetro cambiado
Go = 90
# Generar la ecuación con el nuevo parámetro
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
 DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
 return DA_K
DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# 45°
a = 2.5
def L_45(a, Y):
 L_45 = a*Y
 return L_45
L_45 = L_45(a, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #curva ORIGINAL
ax.plot(Y, DA_G, label = "DA_G", color = "#EE7600", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "#404040") #línea de 45º

# Lineas punteadas
plt.axvline(x = 75, ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axhline(y = 188, xmin= 0, xmax = 0.75, linestyle = ":", color = "grey")
plt.axvline(x = 56, ymin= 0, ymax = 0.56, linestyle = ":", color = "grey")
plt.axhline(y = 139, xmin= 0, xmax = 0.55, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 150, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 200, '$DA_G$', fontsize = 11.5, color = '#EE7600')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(78, 0, '$Y_G$', fontsize = 12, color = '#EE7600')
plt.text(60, 45, '$→$', fontsize = 18, color = 'grey')
plt.text(20, 140, '$↑$', fontsize = 18, color = 'grey')

# Título y leyenda
ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')
ax.legend()
plt.show()
```

### ¿Cuál es el papel que juega el parametro  en el multiplicador keynesiano?
En primer lugar el multiplicador keynesiano es un cambio que ocurre entre los componentes de la demanda agregada (DA), lo cual acaba formando un nuevo producto de equilibrio.
El papel del parametro es que "el tamano del multiplicador keynesiano depende de la magnitud de los parametros" (Jimenez, F. "Apuntes de Macroeconomia")
